here are five plots that we can plot using the Seaborn library, along with their uses:

Scatter plot: A scatter plot is a two-dimensional plot that displays the relationship between two variables. It is useful for visualizing patterns and correlations between variables.

Line plot: A line plot is a graph that displays data as points that are connected by lines. It is useful for visualizing trends over time.

Histogram: A histogram is a graph that displays the distribution of a numerical variable. It is useful for understanding the shape and spread of a dataset.

Bar plot: A bar plot is a graph that displays categorical data with rectangular bars. It is useful for comparing the values of different categories.

Heatmap: A heatmap is a two-dimensional plot that displays the relationships between two categorical variables by color-coding the cells of a grid. It is useful for identifying patterns and relationships between categorical variables.
