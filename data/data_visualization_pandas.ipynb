```python
my_dict={'height':[6,6.5,7,5,5.2],'weight':[90,95,98,70,72],'gender':['male','male','male','female','female']}
```

```python
df=pd.DataFrame(my_dict)
```

```python
df.head()
```

```python
df['color']=df['gender'].map({'male':'red','female':'blue'})
```

```python
df.head()
```

```python
df.plot.scatter(y='height',x='weight',c='color')
```

```python
ax=df.plot(y='height',x='weight',kind='scatter',c='red')
df.plot(y='height',x='weight',kind='scatter',c='blue',ax=ax)
```
