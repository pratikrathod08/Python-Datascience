```python
#  displot methods
```

```python
penguins = sns.load_dataset('penguins')
sns.displot(data=penguins,x='flipper_length_mm')
```

```python
sns.displot(data=penguins,x='flipper_length_mm',kde=True)
```

```python
sns.displot(data=penguins,x='flipper_length_mm',kde=True,hue='species')
```

```python
sns.displot(data=penguins,x='flipper_length_mm',hue='species',kind='kde')
```

```python
sns.displot(data=penguins,x='flipper_length_mm',hue='species',col='sex',kind='kde')
```

```python
#  plot a univariate analysis along with x axis
```

```python
sns.ecdfplot(data=penguins,x='flipper_length_mm')
```

```python
sns.ecdfplot(data=penguins,y='flipper_length_mm')
```

```python
sns.ecdfplot(data=penguins,x='bill_length_mm',hue='species')
```

```python
sns.ecdfplot(data=penguins,x='bill_length_mm',hue='species',stat='count')
```

#  Histogram

```python
sns.set(style='darkgrid',font_scale=1.2)
sns.displot(data=penguins,x='bill_length_mm',kind='hist',aspect=1.4,log_scale=10)
```

```python
sns.displot(data=penguins,x='bill_length_mm',kind='hist',aspect=1.4,log_scale=10,bins=20)
```

```python
sns.displot(data=penguins,x='bill_length_mm',hue='species',kind='hist',aspect=1.4,log_scale=10,bins=20)
```

```python
sns.displot(data=penguins,x='bill_length_mm',col='species',aspect=1.4,log_scale=10,bins=20)
```

```python
sns.displot(data=penguins,x='bill_length_mm',y='bill_depth_mm',col='species',kind='hist',height=5,aspect=1.2,log_scale=(10,0),bins=20)
```

#  Stripplot

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(data=tips,x='tip',y='day')
plt.show()
```

```python
sns.stripplot(data=tips,x='tip',y='day')
plt.show()
```

```python
sns.stripplot(data=tips,x='tip',y='day')
plt.show()
```

```python
sns.stripplot(data=tips,x='tip',y='day',hue='day',jitter=True)
plt.show()
```

```python
sns.stripplot(data=tips,x='tip',y='day',hue='day',jitter=True,alpha=0.5)
plt.show()
```

# Joint plot

```python
sns.jointplot(x='size',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

#  Swarmplot

```python
tips.head()
```

```python
plt.figure(figsize=(11,6))
sns.swarmplot(x='day',y='total_bill',data=tips)
```

#  Violine plot

```python
sns.set(style='whitegrid')
sns.violinplot(data=tips,x='total_bill',orient='h')
plt.title('Total_bill')
```

```python
sns.set(style='whitegrid')
sns.violinplot(data=tips,y='sex',x='tip')
plt.title('Tips')
plt.show()
```

```python
# split violine plot
```

```python
sns.violinplot(data=tips,x='tip',y='sex',hue='smoker',split=True)
plt.title('Tips')
plt.show()
```
