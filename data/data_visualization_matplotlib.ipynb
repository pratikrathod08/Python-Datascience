#  Nested Pie Chart

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
source_data={'students':['jake','amy','boyle'],
            'math_score':[68,82,97],
            'english_score':[70,93,95],
            'phsics_score':[73,85,95]}
```

```python
df=pd.DataFrame(source_data)
```

```python
df
```

```python
df['cummulative_score']=df['math_score']+df['english_score']+df['phsics_score']
```

```python
df
```

```python
x1=df.iloc[0:3,1]
x2=df.iloc[0:3,2]
```

```python
x1
```

```python
x2
```

```python
cmap=plt.get_cmap('tab20c')
outer_colors=cmap(np.arange(3)*4)
inner_colors=cmap(np.array([1,5,9]))
```

```python
#  setting the size of figure

plt.figure(figsize=(8,6))

# plotting the outer pie

plt.pie(x1,labels=df.iloc[0:3,0],startangle=90,pctdistance=0.88,colors=outer_colors,autopct='%1.1f%%',
       radius=1.0,labeldistance=1.05,
       textprops={'fontweight':'bold','fontsize':13},
       wedgeprops={'linewidth':3,'edgecolor':'w'})

# plotting the inner pie

plt.pie(x2,startangle=90,pctdistance=0.85,colors=inner_colors,
       autopct='%1.1f%%',radius=0.60,
       textprops={'fontweight':'bold','fontsize':13},
       wedgeprops={'linewidth':3,'edgecolor':'w'})

center_circle=plt.Circle((0,0),0.25,fc='white')
fig=plt.gcf()
fig.gca().add_artist(center_circle)

#  plotting the pie

plt.axis('equal')
plt.legend(loc=3,fontsize=15)
plt.tight_layout()
plt.show()
```
